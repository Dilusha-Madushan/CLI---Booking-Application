```python
def scrape_transactions(start_date, end_date):
    # Convert the date strings to datetime objects for comparison
    start_date = datetime.strptime(start_date, '%Y%m%d')
    end_date = datetime.strptime(end_date, '%Y%m%d')

    # Define the download path and ensure the directory exists
    download_path = '/content/drive/MyDrive/Colab Notebooks/FYP/bitcoin trx data/'
    if not os.path.exists(download_path):
        os.makedirs(download_path)

    # Set up Chrome options
    chrome_options = Options()
    chrome_options.add_argument('--headless')
    chrome_options.add_argument('--no-sandbox')
    chrome_options.add_argument('--disable-dev-shm-usage')
    prefs = {
        "download.default_directory": download_path,
        "download.prompt_for_download": False,
    }
    chrome_options.add_experimental_option("prefs", prefs)
    # URL of the transactions page
    url = "http://blockdata.loyce.club/transactions/"

    # Start a new instance of the browser with the custom download path
    # browser = webdriver.Chrome(executable_path='chromedriver', options=chrome_options)
    browser = webdriver.Chrome(options=chrome_options)



    # Navigate to the URL
    browser.get(url)
    time.sleep(3)  # wait for 3 seconds to ensure the page is loaded

    # Extract all the links that end with .tsv.gz
    links = browser.find_elements(By.XPATH, "//a[contains(@href, '.tsv.gz')]")

    # Download files within the specified date range
    for link in links:
        file_name = link.get_attribute("text")
        file_date_str = file_name.split('_')[-1].split('.')[0]  # Extract the date part from the file name
        file_date = datetime.strptime(file_date_str, '%Y%m%d')

        # Check if the file date is within the specified range
        if start_date <= file_date <= end_date:
            file_url = link.get_attribute("href")
            browser.get(file_url)
            print("got the file", file_url)
            time.sleep(5)  # wait for 5 seconds to ensure the file is downloaded

    # Close the browser
    browser.quit()
```

# Extract Tranaction, Input trx then combine

```python
import os
import pandas as pd
from datetime import datetime, timedelta

def process_files(start_date, end_date):
    base_url = "http://blockdata.loyce.club"
    download_path = "/content/drive/MyDrive/Colab Notebooks/FYP/data"  # Replace with the path to your directory
    # Convert date strings to datetime objects
    start_date = datetime.strptime(start_date, '%Y%m%d')
    end_date = datetime.strptime(end_date, '%Y%m%d')

    current_date = start_date
    while current_date <= end_date:
        date_str = current_date.strftime('%Y%m%d')

        # Define URLs for the input transactions and transactions data
        input_url = f"{base_url}/inputs/blockchair_bitcoin_inputs_{date_str}.tsv.gz"
        trx_url = f"{base_url}/transactions/blockchair_bitcoin_transactions_{date_str}.tsv.gz"

        # Download the data
        df_input = pd.read_csv(input_url, compression='gzip', sep='\t')
        df_trx = pd.read_csv(trx_url, compression='gzip', sep='\t')

        # Filter and rename columns in the input transactions data
        wanted = ["recipient", "spending_transaction_hash", "time", "value"]
        df_input_filtered = df_input[wanted]
        df_input_filtered = df_input_filtered.rename(columns={"recipient": "account", "spending_transaction_hash": "hash"})

        # Merge the dataframes
        df_merged = pd.merge(df_input_filtered, df_trx, on='hash', how='inner')

        # Define the save paths and ensure the directories exist
        input_folder = os.path.join(download_path, 'inputs')
        trx_folder = os.path.join(download_path,  'transactions')
        merged_folder = os.path.join(download_path, 'merged')

        for folder in [input_folder, trx_folder, merged_folder]:
            if not os.path.exists(folder):
                os.makedirs(folder)

        # df_input.to_csv(os.path.join(input_folder, f"inputs_{date_str}.tsv.gz"), index=False, sep='\t', compression='gzip')
        # df_trx.to_csv(os.path.join(trx_folder, f"transactions_{date_str}.tsv.gz"), index=False, sep='\t', compression='gzip')
        df_merged.to_csv(os.path.join(merged_folder, f"merged_{date_str}.tsv.gz"), index=False, sep='\t', compression='gzip')

        print(f"Processed and saved data for {date_str}")

        # Move to the next date
        current_date += timedelta(days=1)
```

# File delete function

```python
import os
from datetime import datetime, timedelta

def delete_files_in_range(start_date, end_date, directory_path):
    # Convert date strings to datetime objects
    start_date = datetime.strptime(start_date, '%Y%m%d')
    end_date = datetime.strptime(end_date, '%Y%m%d')

    current_date = start_date
    while current_date <= end_date:
        # Generate the file name based on the date
        file_name = f"blockchair_bitcoin_transactions_{current_date.strftime('%Y%m%d')}.tsv.gz"
        file_path = os.path.join(directory_path, file_name)

        # Check if the file exists and delete it
        if os.path.exists(file_path):
            os.remove(file_path)
            print(f"Deleted: {file_path}")
        else:
            print(f"File not found: {file_path}")

        # Move to the next date
        current_date += timedelta(days=1)

# Example usage:
directory_path = "/content/drive/MyDrive/Colab Notebooks/FYP/bitcoin trx data"  # Replace with the path to your directory
delete_files_in_range('20200526', '20210526', directory_path)
```

```python
import os
import pandas as pd

def generate_sql_file(start_date, end_date, output_file):

    # File paths
    file_prefix = "/content/drive/MyDrive/Colab Notebooks/FYP/data/merged/merged_"

    # Generate dates in the range
    dates = pd.date_range(start=start_date, end=end_date, freq='D').strftime('%Y%m%d').tolist()

    # Open the output file for writing
    with open(output_file, 'w') as f:

        # Iterate through each file
        for date in dates:
            file_path = f"{file_prefix}{date}.tsv.gz"

            # Check if file exists
            if not os.path.exists(file_path):
                print(f"File for date {date} does not exist. Skipping...")
                continue

            # Read the file
            df = pd.read_csv(file_path, compression='gzip', header=0, sep='\t', quotechar='"')

            # Convert time_y to datetime
            df['time'] = pd.to_datetime(df['time_y'])

            # Convert value to int (or use appropriate conversion based on the nature of your data)
            df['value'] = df['value'].astype(int)

            # Convert the datetime column to UNIX timestamp
            df['time'] = df['time'].astype(int) / 10**9

            # Filter necessary columns
            df_filtered = df[['account', 'time', 'value']]

            # Write the INSERT statements to the output file
            for index, row in df_filtered.iterrows():
                query = f"INSERT INTO account_time_value (account, time, value) VALUES ('{row['account']}', '{row['time']}', '{row['value']}');\n"
                f.write(query)

    print(f"SQL file generated: {output_file}")
```

```python
import os
import pandas as pd

def extract_records_by_accounts(start_date, end_date, accounts_list):

    # File paths
    file_prefix = "/content/drive/MyDrive/Colab Notebooks/FYP/data/acc details csv/acc_details_"
    output_path = "/content/drive/MyDrive/Colab Notebooks/FYP/data/extracted/"

    # Create output directory if it doesn't exist
    if not os.path.exists(output_path):
        os.makedirs(output_path)

    # Generate dates in the range
    dates = pd.date_range(start=start_date, end=end_date, freq='D').strftime('%Y%m%d').tolist()

    extracted_records = []

    # Iterate through each file
    for date in dates:
        file_path = f"{file_prefix}{date}.tsv.gz"

        # Check if file exists
        if not os.path.exists(file_path):
            print(f"File for date {date} does not exist. Skipping...")
            continue

        # Read the file
        df = pd.read_csv(file_path, compression='gzip', header=0, sep='\t', quotechar='"')

        # Filter rows by accounts
        df_extracted = df[df['account'].isin(accounts_list)]

        # Append to the list
        extracted_records.append(df_extracted)

    # Concatenate all extracted records
    final_df = pd.concat(extracted_records, ignore_index=True)

    # Save the extracted records to a new CSV
    final_df.to_csv(os.path.join(output_path, "extracted_records.csv"), index=False)

    print("Extraction complete!")
```
